```python
import pandas as pd

url_data = ('https://raw.githubusercontent.com/ai2-education-fiep-turma-5/02-programacao-python/main/src/aula11/data/aluguel.csv')
df = pd.read_csv(url_data, sep = ';') 

```

```python
class Database:

    def __init__(self, local):
        self.local = local
        self.df = pd.read_csv(self.local, sep = ';')

    def leitura(self):
        #Lê o arquivo

        return self.df

    def agrupamentos(self, coluna):
        #Agrupa de acordo com a coluna e retorna a media dos valores
        try:
            self.coluna = coluna    
            return self.df.groupby([str(self.coluna)]).mean()
        except Exception:
            print('Essa coluna não existe!')

    def troca(self,coluna,index,quanto):
        #Troca um valor do dataframe por outro

        self.coluna = coluna
        self.index = index
        self.quanto = quanto
        try:
            self.anterior = self.df[self.coluna][self.index]
            self.df[self.coluna][self.index] = self.quanto
            print('O valor da coluna {} na posição {} foi trocado de {} para {}'.format(self.coluna, self.index, self.anterior, self.quanto))
            return self.df
        except Exception:
            print('Essa posição não existe no dataframe.')

    def inserirlinha(self):
        #Insere uma nova linha (deve ser um dicionário)

        self.tipo = input('Qual o tipo?')
        self.bairro = input('Qual o bairro?')
        self.quartos = input('Quantos quartos?')
        self.vagas = input('Quantas vagas?')
        self.suites = input('Quantas suites?')
        self.area = input('Qual a area?')
        self.valor = input('Qual o valor?')
        self.condo = input('Quanto é o condomínio?')
        self.iptu = input('Quanto é o IPTU?')
        self.df = self.df.append({'Tipo':self.tipo,'Bairro':self.bairro,'Quartos':self.quartos,'Vagas': self.vagas, 'Suites': self.suites, 'Area': self.area, 
        'Valor': self.valor, 'Condominio': self.condo, 'IPTU': self.iptu}, ignore_index = True)
        return self.df
    
    def excluirlinha(self,linha):

        #Troca um valor do dataframe por outro
        try:
            self.linha = linha
            self.df = self.df.drop(self.linha)
            print('A linha {} foi excluída.'.format(self.linha))
            return self.df
        except Exception:
            print('Essa linha não existe no dataframe.')
            
    def salvar(self,nome):
        #Salva o dataframe como csv
        
        self.nome = nome
        self.df.to_csv(nome + '.csv', index = False)
        print('O arquivo foi salvo como:',nome + '.csv','.')

```

### 1.1 leitura da base de dados;

```python
data = Database(url_data)
data.leitura()
```

### 1.2 agrupamentos:

```python
data.agrupamentos('Cachorro')
```

```python
data.agrupamentos('Tipo')
```

### 1.3 troca de valores:

```python
data.troca('Cachorro',1,'Hub de IA')
```

```python
data.troca('Tipo',1,'Hub de IA')
```

### 1.4 inserção de valores (Aqui os valores são inseridos através do input);

```python
data.inserirlinha()
```

### 1.5 exclusão de valores;

```python
data.excluirlinha(2)
```

### 1.6 salvar um `df`

```python
data.salvar('Dados')
```
